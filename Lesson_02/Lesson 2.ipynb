## ДЗ Урок 2. Работа с библиотекой Pandas, часть 1. Загрузка данных и анализ их структуры

1. Загрузить, просмотреть, определить количество строк и склеить 3 датасета: marketing_campaign.csv, users.csv и subscribers.csv.

```python
import pandas as pd
import numpy as np

df_mc = pd.read_csv('marketing_campaign.csv')
df_mc.info()
df_mc
```

```python
df_users = pd.read_csv('users.csv')
df_users.info()
df_users
```

```python
df_subscribers = pd.read_csv('subscribers.csv')
df_subscribers.info()
df_subscribers
```

```python
df = df_mc.merge(df_users, on = 'user_id').merge(df_subscribers, on = 'user_id')
df
```

2. Определить количество, типы и описательные статистики колонок (столбцов) получившегося датасета.

```python
df.info()
df.describe()
```

3. Определить эффективность маркетинговых каналов по привлечению платящих игроков.

```python
df[df['converted'] == True]['subscribing_channel'].value_counts(normalize = True) * 100
```

4. Определить самую раннюю дату подписки на сервис.

```python
df[df['date_subscribed'].str.contains('nan') == False]['date_subscribed'].min()
```

5. Определить портрет аудитории удержанных подписчиков (по возрасту и языку).

```python
df[df['is_retained'] == True].filter(items = ['age_group', 'language_preferred']).value_counts()
```
